```python
# Importing Train and Test Datasets
train1=pd.read_csv("Train_Data_X.csv")
```

```python
train2=pd.read_csv("Train_Data_Y.csv")
```

```python
test=pd.read_csv("Test_Data_X.csv")
```

```python
# Dropping the "Unnamed:0" column for convenience
train1=train1.drop("Unnamed: 0",axis=1)
```

```python
train2=train2.drop("Unnamed: 0",axis=1)
```

```python
train1.head()
```

```python
train2.head()
```

```python
# Converting "train1" and "train2" to dataframes inorder to combine them into a single dataset by the name "train"
df1=pd.DataFrame(train1)
```

```python
df2=pd.DataFrame(train2)
```

```python
train=pd.concat([df1,df2],axis=1,sort=False)
```

```python
train
```

```python
train.columns
```

```python
predictor_vars=['X_follower_count', 'X_following_count', 'X_listed_count',
       'X_mentions_received', 'X_retweets_received', 'X_mentions_sent',
       'X_retweets_sent', 'X_posts', 'X_network_feature_1',
       'X_network_feature_2', 'X_network_feature_3', 'Y_follower_count',
       'Y_following_count', 'Y_listed_count', 'Y_mentions_received',
       'Y_retweets_received', 'Y_mentions_sent', 'Y_retweets_sent', 'Y_posts',
       'Y_network_feature_1', 'Y_network_feature_2', 'Y_network_feature_3']
```

```python
x,y=train[predictor_vars],train[['X_social_than_Y?']]
```

```python
x.head()
```

```python
y.head()
```

```python
# Splitting the datasets for efficient implementation of ML Algorithm
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
# the test and train datasets from the 'train' dataset are in the ratio 25:75
len(X_train)
```

```python
len(X_test)
```

```python
# Implementing Logistic Regression model
from sklearn.linear_model import LogisticRegression

logReg = LogisticRegression()

logReg.fit(X_train, y_train)
```

```python
logReg_y_pred = logReg.predict(X_test)
```

```python
# From the output of this confusion matrix we can see that the Logistic Regression model predicted 265-TN ,86-FP, 93-FN and 234-TP 
# In this case P means A is more influential than B, N means B is more influential than A
# So, we can say that this algorithm predicted more True values than False values

from sklearn.metrics import confusion_matrix,accuracy_score, auc, f1_score
logReg_CM = confusion_matrix(y_test, logReg_y_pred)
logReg_CM
```

```python
logReg_accuracy = accuracy_score(y_test,logReg_y_pred)
logReg_accuracy
```

```python
# f1 score was calculated inspite of calculating Accuracy score 
# Because Accuracy is the most intuitive performance measure and it is simply a ratio of correctly predicted observation to the total observations.
# F1 Score is the weighted average of Precision and Recall
# The f1 score obtained was 0.72

logReg_f1 = f1_score(y_test,logReg_y_pred)
logReg_f1
```

```python
# For fine tuning the prediction accuracy, Random Forest Classification Model was implemented

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc
```

```python
rfc.fit(x,y)
rfc_y_pred = rfc.predict(X_test)
```

```python
accuracy_score(y_test,rfc_y_pred)
```

```python
rfc_f1 = f1_score(y_test,rfc_y_pred)
rfc_f1
```

```python
# The obtained values of Accuracy score and f1 score were high enough to implement this Algorithm 
rfc_y_pred
```

```python
# It can be seen that the predicted values are almost similar to the true values

y_test
```

```python
# The Random Forest Classification ML Algorithm was implemented on the 'test' dataset

predictions=rfc.predict(test[predictor_vars])
```
